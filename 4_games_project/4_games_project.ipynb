# Изучение закономерностей, определяющих успешность игр

**Описание проекта**

Объект исследования - интернет-магазин «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Имеются данные до 2016 года. Представим, что сейчас декабрь 2016 г., и планируется кампания на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируем ли мы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

### Обработка пропусков

Выведем количество пропущенных значений и их процент от общего. Нас интересуют значения `Name, Year_of_Release, Genre, Critic_Score, User_Score, Rating`.

```python
data_isna = pd.DataFrame()
data_isna['sum'] = data.isna().sum() 
data_isna['percent'] = (data.isna().mean()*100).round(1).astype(str).add('%')
data_isna
```

Пропусков в названии `name`, году выпуска `year_of_release` и жанре игры `genre` мало, поэтому просто отфильтруем строки с пропусками этих значений. Их удаление не повлияет на результаты исследования.

```python
#data.dropna(subset = ['name', 'year_of_release', 'genre'], inplace=True)

data = data.query('not name.isna() and not year_of_release.isna() and not genre.isna()')
```

В столбце `user_score` есть абревиатура `tbd` (подлежит уточнению). Обработаем его заменой на `NaN`, так как по сути это пропущенное значение.

```python
data.loc[data['user_score'] == 'tbd', 'user_score'] = float('nan')
```

Посмотрим сколько пропусков одновременно и оценок критиков, и оценок пользователей, и рейтинга.

```python
len(data.loc[data['critic_score'].isna() & data['user_score'].isna() & data['rating'].isna()])
```

Строк без какой либо оценки - 6580. 

Пропуски в стролбцах `critic_score`, `user_score` и `rating` оставим без изменений. Удаление не рассматривается, так как их слишком много. А замены не возможны так как в большинстве случаев нет ни одной из оценок, они исказят результаты корреляционного анализа и стат-тестов.

### Исследование самых популярных платформ

Посмотрим, как менялись продажи по платформам. С помощью `value_counts()` выделим платформы с наибольшими суммарными продажами и сохраним их в список `top_10_platforms`. 

```python
data.groupby('platform').agg({'total_sales': 'sum'})\
                        .sort_values(by='total_sales', ascending=False).head(10)
```

```python
top_10_platforms = ['PS2', 'X360', 'Wii', 'PS3', 'DS', 'PS', 'GBA', 'PS4', 'PSP', '3DS']
```

Создадим сводную таблицу, где строки будут годы, столбцы платформы, которые входят в топ-10, а значения общие продажи `total_sales`.

Построим график по полученным данным.

```python
data_sales_per_year = (data
    .query('platform in @top_10_platforms')
    .pivot_table(index='year_of_release', columns='platform', values='total_sales', aggfunc='sum')
)

data_sales_per_year.plot(figsize=(23,8), style='--')
plt.title('Продажи по топ-10 платформам', size=20)
plt.xlabel('Год', size=16)
plt.ylabel('Количество продаж в млн копий', size=16)
plt.legend()
plt.grid()
plt.show()
```

На графике видно, что полулярность платформ похожа на горы, они сначала набирают популярность, потом теряют. Бывает один или несколько пиков. Какие-то дольше набирают популярность и быстро исзезают, какие-то наоборот.

### Выбор актуального периода  и потенциально прибыльных платформ

Чтобы получить точное значение среднего срока жизни платформ, используем сводные таблицы. Рассмотрим продажи больше 5 млн копий.

```python
min_sales = 5

platform_lifetime = data.pivot_table(index = ['platform', 'year_of_release'], 
                                      values = 'total_sales', aggfunc = 'sum')\
                                    .query('total_sales > @min_sales')

platform_lifetime = platform_lifetime.reset_index().pivot_table(index = 'platform', 
                                                                values = 'year_of_release', 
                                                                aggfunc = 'nunique').mean()

print('Средний срок существования платформы:', platform_lifetime['year_of_release'].mean())
```

Чтобы построить прогноз на 2017 год, возьмем половину среднего срока существования платформы - 4 года. Так как половину жизненного цикла платформа растет, а вторую половину падает. Значит, актуальным периодом будем считать данные с 2013 года.

Отвильтруем данные и выведем самые прибыльные платформы за последние 4 года.

```python
new_data = data.query('year_of_release >= 2013')

new_data.groupby('platform').agg({'total_sales': 'sum'})\
                        .sort_values(by='total_sales', ascending=False).head(10)
```

Топ платформ уже другой. На первом месте теперь `PS4`, далее `PS3` и `XOne`.

Еще раз построим график по данным актуального периода.

```python
data_sales_per_year = new_data.pivot_table(index='year_of_release', columns='platform', 
                                           values='total_sales', aggfunc='sum')

data_sales_per_year.plot(figsize=(23,8), style='--')
plt.title('Продажи по актуальному периоду', size=20)
plt.xlabel('Год', size=16)
plt.ylabel('Количество продаж в млн копий', size=16)
plt.legend()
plt.grid()
plt.show()
```

## Проверка гипотез

### Гипотеза пользовательские рейтинги платформ Xbox One и PC одинаковые

Проверим гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Для проверки импользуем  t-тест для гипотез:
* H0: средние пользовательские рейтинги платформ Xbox One и PC равны.
* H1: средние пользовательские рейтинги платформ Xbox One и PC не равны

Используем метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам `scipy.stats.ttest_ind` с параметрами рейтинга пользователей `user_score` по платформе `platform`.

Проверим дисперсии выборок, чтобы корректно определить атрибут `equal_val` в Т–тесте

```python
# оценки платформы X One без пропущенных значений
xone_user_score = new_data.query('platform == "XOne"')['user_score'].dropna()
# оценки платформы PC без пропущенных значений
pc_user_score = new_data.query('platform == "PC"')['user_score'].dropna()

# сравнение дисперсий
print(np.var(xone_user_score), np.var(pc_user_score))
```

Дисперисии сильно отличаются, поэтому при проведение t-теста параметр `equal_var` примем как `False`.

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвегнем гипотезу

results = st.ttest_ind(xone_user_score, pc_user_score, equal_var = False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Нулевую гипотезу не получилось отвергнуть. Вероятность равенства средних пользовательских рейтингов платформ Xbox One и PC есть.

Для наглядности построим графики распределения оценок пользователей по платформам XOne и PC.

```python
xone_user_score.plot(kind='kde', figsize=(10,5), label="XOne")
pc_user_score.plot(kind='kde', figsize=(10,5), label="PC")
plt.title('Пользовательские рейтенги XOne и PC')
plt.xlabel('Оценка')
plt.legend()
plt.show()
```

### Гипотеза пользовательские рейтинги жанров Action и Sports разные

Проверим гипотезу о том, что средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Для проверки импользуем  t-тест для следующих статистических гипотез:
* H0: средние пользовательские рейтинги жанров Action и Sports равны
* H1: средние пользовательские рейтинги жанров Action и Sports не равны

Используем метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам `scipy.stats.ttest_ind` с параметрами рейтинга пользователей `user_score` по жанру `genre`.

```python
# оценки жанра Action без пропущенных значений
action_user_score = new_data.query('genre == "Action"')['user_score'].dropna()
# оценки жанра Sports без пропущенных значений
sports_user_score = new_data.query('genre == "Sports"')['user_score'].dropna()

# сравнение дисперсий
print(np.var(xone_user_score), np.var(pc_user_score))
```

Дисперисии сильно отличаются, поэтому при проведение t-теста параметр `equal_var` примем как `False`.

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвегнем гипотезу

results = st.ttest_ind(action_user_score, sports_user_score, equal_var = False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

По результату t-теста нулевую гипотезу отвергаем. Следовательно альтернативную гипотезу о том, что средние пользовательские рейтинги жанров Action и Sports разные отвергнуть мы не можем. Это вероятно.

Для наглядности построим график.

```python
action_user_score.plot(kind='kde', figsize=(10,5), label="XOne")
sports_user_score.plot(kind='kde', figsize=(10,5), label="PC")
plt.title('Пользовательские рейтенги жанров Action и Sports')
plt.xlabel('Оценка')
plt.legend()
plt.show()
```

### Вывод

В результате проверки с помощью t-теста и графика распределения, гипотезы не получилось отвергнуть. То есть гипотеза о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые, и гипотеза о том, что средние пользовательские рейтинги жанров Action и Sports разные, вероятны.
